# Part1: 

### (a) upside-down lena.bmp

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt

originalImage = cv2.imread('./lena.bmp', cv2.IMREAD_GRAYSCALE)

Result = originalImage[::-1]#將整個array上下相反

plt.imshow(Result, cmap = 'gray')
cv2.imwrite('a.jpg', Result)
```

### (b) right-side-left lena.bmp

```python
import cv2
import matplotlib.pyplot as plt

originalImage = cv2.imread('./lena.bmp', cv2.IMREAD_GRAYSCALE)

Result = originalImage[:,::-1]#將整個array左右相反

plt.imshow(Result, cmap = 'gray')
cv2.imwrite('b.jpg', Result)
```

### (c) diagonally mirrored lena.bmp

```python
import cv2
import matplotlib.pyplot as plt

Image = cv2.imread('./lena.bmp', cv2.IMREAD_GRAYSCALE)

for i in range(0, Image.shape[0]):
    for j in range(i+1, Image.shape[1]):#每一列主對角線元素以後ij做swap
        Image[i][j],Image[j][i] = Image[j][i],Image[i][j]
        
plt.imshow(Image, cmap = 'gray')
cv2.imwrite('c.jpg', Image)
```

# Part2: 

### (f) binarize lena.bmp at 128 to get a binary image

```python
import cv2
import matplotlib.pyplot as plt

Image = cv2.imread('./lena.bmp', cv2.IMREAD_GRAYSCALE)

for i in range(0, Image.shape[0]):
    for j in range(0, Image.shape[1]):#超過128變255 否則變0
        if(Image[i][j]>128):
            Image[i][j] = 255
        else:
            Image[i][j] = 0
plt.imshow(Image, cmap = 'gray')
cv2.imwrite('f.jpg', Image)
```
