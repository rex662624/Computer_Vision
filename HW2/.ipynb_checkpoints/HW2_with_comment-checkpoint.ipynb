```python
#Use Local Equivalence Table

import threading,time

st = time.clock()
CCL_label = 2
last_pixel = []
ifrun = True
runtime = 0
def connected_4(lena_b,x,y):
    con_min = lena_b[x,y]
    if lena_b[x-1,y+1] < con_min and lena_b[x-1,y+1] != 0 and lena_b[x-1,y+1] != 255:
        con_min = lena_b[x-1,y+1]
    if lena_b[x,y+1] < con_min and lena_b[x,y+1] != 0 and lena_b[x,y+1] != 255:
        con_min = lena_b[x,y+1]
    if lena_b[x+1,y+1] < con_min and lena_b[x+1,y+1] != 0 and lena_b[x+1,y+1] != 255:
        con_min = lena_b[x+1,y+1]
    if lena_b[x-1,y] < con_min and lena_b[x-1,y] != 0 and lena_b[x-1,y] != 255:
        con_min = lena_b[x-1,y]
    if lena_b[x+1,y] < con_min and lena_b[x+1,y] != 0 and lena_b[x+1,y] != 255:
        con_min = lena_b[x+1,y]
    if lena_b[x-1,y-1] < con_min and lena_b[x-1,y-1] != 0  and lena_b[x-1,y-1] != 255:
        con_min = lena_b[x-1,y-1]
    if lena_b[x,y-1] < con_min and lena_b[x,y-1] != 0 and lena_b[x,y-1] != 255:
        con_min = lena_b[x,y-1]
    if lena_b[x+1,y-1] < con_min and lena_b[x+1,y-1] != 0 and lena_b[x+1,y-1] != 255:
        con_min = lena_b[x+1,y-1]
    """if con_min == 1:
        print("inner connect_check: x,y",x,y," conected:", lena_b[x,y],',',
              lena_b[x,y+1],',',lena_b[x-1,y],',',lena_b[x+1,y],',',lena_b[x,y-1],',',con_min)"""
        
    return con_min 
def traceback(last_pixel,temp_min):
    #print(last_pixel,',',temp_min)
    for x in range(len(last_pixel)):
        #print("x: ",x,',',last_pixel[x][0],last_pixel[x][1])
        lena_ccl[last_pixel[x][0],last_pixel[x][1]] = temp_min
lena_ccl = lena_b.astype('uint16')
#@jit(nopython=True)
def CCL():
    global ifrun,CCL_label,last_pixel,runtime,st
    while ifrun:
        ifrun = False
        #print("top_down")
        for x in range(1,rows-1):
            last_pixel = []
            for y in range(1,cols-1):              
                if lena_ccl[x,y] == 255:
                    #print("x,y,lena_ccl: ",x,",",y,",",lena_ccl[x,y])
                    temp_min = connected_4(lena_ccl,x,y)
                        #print("x,y,min: ",x,",",y,",",temp_min)
                    if temp_min != 255:
                        #print("find old label: ", 'temp_min:',temp_min,'CCL_label:',CCL_label,'last: ', lena_ccl[x,y-1])
                        lena_ccl[x,y] = temp_min
                        last_pixel.append([x,y])
                        if lena_ccl[x,y-1] !=0 and lena_ccl[x,y-1] > temp_min:
                            #print("\033[33m traceback: \033[0m")
                            traceback(last_pixel,temp_min)
                            ifrun = True
                    else:
                        #print("\033[35m give new label: \033[0m", CCL_label)
                        lena_ccl[x,y] = CCL_label
                        if lena_ccl[x,y] == 1:
                            print("give 1  ccl label",CCL_label, "lena_ccl[x,y]",lena_ccl[x,y])
                        CCL_label = CCL_label + 1
                        if CCL_label == 255:
                            CCL_label = CCL_label + 1
                        ifrun = True
                else:
                    last_pixel = []
                #if lena_ccl[x,y] == 1:
                    #print("\033[33m lena_CCL: \033[0m",lena_ccl[x,y], "temp_min", temp_min," x,y",x,y)
        #print("Buttom_up")
        for x in range(rows-1-1,0,-1):
            for y in range(cols-1-1,0,-1):
                if lena_ccl[x,y] != 0:
                    last_pixel.append([x,y])
                    temp_min = connected_4(lena_ccl,x,y)
                    #print('x,y: ',x,',',y,', ',lena_ccl[x,y], temp_min, 'last_pixel: ', last_pixel)
                    if temp_min < lena_ccl[x,y]:
                        #print("\033[33m update: \033[0m")
                        traceback(last_pixel, temp_min)
                        ifrun = True
                else:
                    last_pixel = []
        #print('ifrun:',ifrun)
        #print('runtime:',runtime,"  #",duration,round(duration/60, 2),round(duration/3600, 2),end="\r")
        print('runtime:',runtime, end="\r")
        runtime = runtime + 1

CCL()
np.set_printoptions(edgeitems=512)
#%timeit CCL()
#print('runtime:',runtime,"  #",duration,round(duration/60, 2),round(duration/3600, 2))
print(lena_ccl)
plt.figure(num=1, figsize=(8,5))
plt.title("lena")
plt.axis('off')
plt.set_cmap('gray')
plt.imshow(lena_ccl)
plt.show()
```
